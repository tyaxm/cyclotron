```python
# 自作モジュールへの変更を自動的にリロードするためのおまじない
%load_ext autoreload
%autoreload 2


import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

import lib.S2Z as stoz
import lib.Z_model as zmodel
import lib.Useful as usf
```

```python
data = np.loadtxt("../data/varicon1_max_1-20M_2026_03_30.s1p", delimiter=" ")
freq, Z = stoz.direct(data)
plt.plot(freq, Z.imag, ".")


popt, pcov = curve_fit(zmodel.C_ideal_imag, freq, Z.imag, p0=[60* 10 ** -12])
plt.plot(freq, zmodel.C_ideal_imag(freq, *popt))

plt.figure()

print(popt)
print(np.sqrt(pcov))
plt.plot(freq, Z.real, ".")

```
