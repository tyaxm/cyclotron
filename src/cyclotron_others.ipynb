取り出しまでに粒子が運動する飛跡を調べる

```python
#電圧の大きさと加速回数、周波数から半径がわかるので、そこから取り出し半径までの粒子の軌跡の長さがわかる

f = 13 #MHz
R = 0.05 #m, 半径
V = 0.9 #kV

N = round(R**2 * 206 * f ** 2 / V)
print("N=", N)

sum_sqrt = 0

for i in range (N):
   sum_sqrt = sum_sqrt + np.sqrt(i+1)
#print(sum_sqrt)

R_all = np.pi * np.sqrt(V / (206 * f ** 2)) * sum_sqrt

print("軌跡長 = ", R_all, "m")
```

平均自由行程

```python
#平均自由行程よりも粒子の軌跡が短くなることが粒子加速の要請である。

k = 1.38 * 10 ** (-23)
T = 300 #K
d = 0.376 #nm
P = 0.001 #Pa
lam = 4 * k * T / (np.pi * d * d * 10 ** (-18) * P)
print("平均自由行程 = ", lam, "m")
```

表皮効果

```python
#表皮効果によって、電流が流れる厚さが制限される

rho = 1.72 * 10 ** (-8) #Ω・m
f = 13 * 10 ** 6 #Hz
u0 = 4 * np.pi * 10 ** (-7)

delta = np.sqrt(rho / np.pi / f / u0) * 10 ** 6 #um
print(delta, "um")
```

```python
#共鳴周波数
L_try = 5 * 10 ** (-6) #H
C_try = 30 * 10 ** (-12) #F
f = 1/ (2 * np.pi) / np.sqrt(L_try * C_try) * 10 ** (-6) #MHz

print(f, "MHz")
```

```python
#減衰の時定数 キャパシタンスを求める

R_try = 200 #kΩ

t_half = np.log(2) * C_try * R_try * 10  ** 6

print(t_half, "us")
```
