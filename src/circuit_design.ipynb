```python
import numpy as np
from scipy import constants
```

実験器具について
- 電磁石の磁束密度
- Dee電極のキャパシタンス
- 真空チェンバーの半径

を測定する。

```python
# INPUT
B         = 0.9   # T
C_Dee_pF  = 35    # pF
r_chamber = 0.075 # m

# 単位変換
C_Dee = C_Dee_pF * 1e-12
```

ここから必要な発信周波数と加速後の粒子のエネルギーが求まる。

```python
q = constants.e
m = constants.proton_mass

omega = q * B / m
f_MHz = omega/ (2 * np.pi) * 1e-6

E = (q * B * r_chamber) ** 2 / (2 * m)
E_keV = E / constants.eV * 1e-3

print("発信周波数      :",f_MHz,"(MHz)")
print("粒子のエネルギー :",E_keV,"(keV)")
```

終端部のLC共振回路になっている。発信周波数とDee電極のキャパシタンスの測定結果から、コイルのインダクタンスを決定できる。

```python
L_2 = 1 / (omega ** 2 * C_Dee)
L_2_uH =  L_2 * 1e6

print("二次コイルのインダクタンス L :", L_2_uH, "(uH)")
```

使う導線の情報とコイルのインダクタンスからコイルを設計する。

以上の情報から終端部が設計できる。ここで終端部について
- Q値
- 寄生抵抗

を測定する。

また、Dee電極への印加電圧の目標値を見積もる。

```python
# INPUT
Q_2 = 1
R_2 = 50 # ohm

V_Dee = 1e3 # Volt
```

Dee電極への目標電圧からパワーアンプの目標出力電力を計算できる。

```python
P_min = V_Dee ** 2 * R_2 * C_Dee / L_2
P = P_min * 2 # 2倍ほど大きめに目標出力電力を定める

print("パワーアンプの目標出力電力 :",P,"(W)")
```
